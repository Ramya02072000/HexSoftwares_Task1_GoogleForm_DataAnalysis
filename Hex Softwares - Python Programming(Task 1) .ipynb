## **Fibonacci Generator**

```python
n = int(input("Enter number of terms: "))

a, b = 0, 1

print("Fibonacci Series:")

for i in range(n):
    print(a, end=" ")
    a, b = b, a + b
```

***Interpretation:***

* This project generates the Fibonacci series using Python.
The program starts with two initial values (0 and 1) and then calculates the next numbers by adding the previous two numbers.

* Based on the user input, the program displays the Fibonacci sequence up to the given number of terms.

* This helps in understanding loops, variables, and basic logic building in Python

# **Game of Tic-Tac-Toe**

```python
import random

board = [" " for _ in range(9)]

def print_board():
    print()
    print(board[0], "|", board[1], "|", board[2])
    print("--+---+--")
    print(board[3], "|", board[4], "|", board[5])
    print("--+---+--")
    print(board[6], "|", board[7], "|", board[8])
    print()

def check_winner(player):
    win_conditions = [
        [0,1,2],[3,4,5],[6,7,8],
        [0,3,6],[1,4,7],[2,5,8],
        [0,4,8],[2,4,6]
    ]
    for condition in win_conditions:
        if board[condition[0]] == board[condition[1]] == board[condition[2]] == player:
            return True
    return False

def player_move():
    move = int(input("Enter position (1-9): ")) - 1
    if board[move] == " ":
        board[move] = "X"
    else:
        print("Invalid move!")
        player_move()

def computer_move():
    move = random.randint(0,8)
    if board[move] == " ":
        board[move] = "O"
    else:
        computer_move()

def play_game():
    print("Tic-Tac-Toe Game Start!")
    print_board()

    for i in range(9):
        if i % 2 == 0:
            player_move()
            if check_winner("X"):
                print_board()
                print("You Win! 🎉")
                return
        else:
            computer_move()
            if check_winner("O"):
                print_board()
                print("Computer Wins! 🤖")
                return

        print_board()

    print("It's a Draw!")

play_game()
```

*Interpretation*

* This project is a simple Tic-Tac-Toe game developed using Python, where a user can play against the computer. The game board is represented using a list, and the player makes a move by entering a position from 1 to 9.

* The computer generates its moves randomly using Python’s random library. After each move, the program checks for a winning condition or a draw.

* This project helps in understanding concepts like loops, conditional statements, functions, and basic game logic implementation in Python.
