# **WEB SCRAPER**

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "http://quotes.toscrape.com/"
response = requests.get(url)

if response.status_code == 200:
    print("Website fetched successfully!")
else:
    print("Failed to fetch website")

soup = BeautifulSoup(response.text, "html.parser")

quotes_data = []
quotes = soup.find_all("div", class_="quote")

for quote in quotes:
    text = quote.find("span", class_="text").text
    author = quote.find("small", class_="author").text
    tags = [tag.text for tag in quote.find_all("a", class_="tag")]

    quotes_data.append({
        "Quote": text,
        "Author": author,
        "Tags": ", ".join(tags)
    })

df = pd.DataFrame(quotes_data)

print(df.head())

df.to_csv("quotes_scraped_data.csv", index=False)
print("Data saved as quotes_scraped_data.csv")
```

**Web Scraper Interpretation**

* Successfully extracted quotes, authors, and tags from the website.
* Data was structured into a clean tabular format using Pandas.
* HTML content was parsed effectively using BeautifulSoup.
* Extracted data was saved as a CSV file for further use.
* Demonstrates practical skills in Python, web scraping, and data handling.

# **EXPENSE TRACKER**

```python
import pandas as pd

expenses = []

while True:
    print("\n--- Expense Tracker ---")
    print("1. Add Expense")
    print("2. View Expenses")
    print("3. Show Summary")
    print("4. Save to CSV")
    print("5. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        date = input("Enter date (YYYY-MM-DD): ")
        category = input("Enter category (Food/Travel/Shopping/Bills/Others): ")
        amount = float(input("Enter amount: "))
        description = input("Enter description: ")

        expenses.append({
            "Date": date,
            "Category": category,
            "Amount": amount,
            "Description": description
        })

        print("Expense added successfully!")

    elif choice == "2":
        if expenses:
            df = pd.DataFrame(expenses)
            print("\nAll Expenses:")
            print(df)
        else:
            print("No expenses recorded yet.")

    elif choice == "3":
        if expenses:
            df = pd.DataFrame(expenses)
            print("\nTotal Expense:", df["Amount"].sum())
            print("\nCategory-wise Summary:")
            print(df.groupby("Category")["Amount"].sum())
        else:
            print("No expenses to summarize.")

    elif choice == "4":
        if expenses:
            df = pd.DataFrame(expenses)
            df.to_csv("expense_tracker.csv", index=False)
            print("Data saved as expense_tracker.csv")
        else:
            print("No expenses to save.")

    elif choice == "5":
        print("Exiting Expense Tracker. Thank you!")
        break

    else:
        print("Invalid choice. Please try again.")
```

# **EXPENSE TRACKER**

* The project allows users to record daily expenses easily.
* Expenses are organized by date, category, amount, and description.
* The program calculates the total expense automatically.
* It provides category-wise summaries for better spending analysis.
* The data is saved in CSV format for future reference.

# **RENT CALCULATOR**

```python
# Rent Calculator

print("---- Rent Calculator ----")

# Step 1 & 2: Input
rent = float(input("Enter total rent: "))
electricity = float(input("Enter electricity bill: "))
water = float(input("Enter water bill: "))
others = float(input("Enter other expenses: "))
people = int(input("Enter number of people: "))

# Step 3: Total expense
total = rent + electricity + water + others

# Step 4: Per person
per_person = total / people

# Step 5: Output
print("\nTotal Expense:", total)
print("Each person should pay:", round(per_person, 2))
```

# **Rent Calculator Interpretation**

* The program collects rent and additional expenses from the user.
* It calculates the total monthly expense accurately.
* The total amount is divided equally among all users.
* Helps in managing shared living costs efficiently.
* Demonstrates basic Python skills like input handling and calculations.
